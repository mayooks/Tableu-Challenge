```python
def get_columns(data):
    columns = data.columns
    is_all_columns_lowercase = all([c.islower() for c in columns])
    if is_all_columns_lowercase:
        return columns
    cleaned_columns = []
    for column in columns:
        column_temp = column.split(" ")
        if len(column_temp) == 2:
            new_column = column.replace(" ", "").lower()
        elif len(column_temp)> 2:
            new_column = column.lower()
        cleaned_columns.append(new_column)
    return cleaned_columns
        
def get_clean_data(list_of_csv_files):
    columns = ['tripduration', 'starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
       'birth year', 'gender']
    
    list_of_pandas_data = []
    skipped_pandas_data = []
    
    for file in list_of_csv_files:
        data = pd.read_csv(file)
        try:
            data.columns = columns
            list_of_pandas_data.append(data)
        except ValueError:
            skipped_pandas_data.append(data)
            print("skipping file - ", file, " due to columns mismatch")
    return (pd.concat(list_of_pandas_data), pd.concat(skipped_pandas_data))
        
```

## Data Formatting and cleanig

```python
bike_data.dtypes
```

```python
bike_data.columns
```

```python
bike_data.rename(columns={'trip duration':'trip duration (seconds)'}, inplace=False,)
#converting starttime, stoptime and birth year to datetime datatype
bike_data['starttime'] =  pd.to_datetime(bike_data['starttime'])
bike_data['stoptime'] =  pd.to_datetime(bike_data['stoptime'])

#Gender (Zero=unknown; 1=male; 2=female). replacing numbers with stringers
bike_data["gender"]= bike_data["gender"].astype(str)
bike_data['gender'].replace({'1': 'Male', '2': 'Female','0':'Unknown'}, inplace=True)

bike_data.dtypes
```

```python
bike_data.head(-5)
```

```python
#inserting year of ride and month of ride columns
bike_data['year_of_ride'] = pd.to_datetime(bike_data['starttime']).dt.year
bike_data['ride_month'] = pd.to_datetime(bike_data['starttime']).dt.month

#calculating rider age at the time of the ride
bike_data["rider_age"] = bike_data['year_of_ride']-bike_data['birth year']
bike_data.head(-5)
```
